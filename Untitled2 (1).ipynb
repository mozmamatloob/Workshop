```python
indicators = [
  
    'NY.GDP.MKTP.CD',   
    'NY.GDP.PCAP.CD',   
    'TX.VAL.MRCH.CD.WT',
    'NE.IMP.GNFS.ZS',   
    'NV.IND.TOTL.ZS',   
    'NV.AGR.TOTL.ZS',   
    'NY.GNS.ICTR.ZS',   
    'FP.CPI.TOTL.ZG',   

    'SP.POP.TOTL',      
    'SP.URB.TOTL.IN.ZS',
    'SM.POP.NETM',     
    'SH.DYN.MORT',      
    'SP.DYN.LE00.IN',   
    'SH.STA.SUIC.P5',  
    'SH.MLR.IMTR.ZS',   

    'SE.ADT.LITR.ZS',   
    'SE.PRM.ENRR',      
    'SE.SEC.ENRR',      
    'SL.UEM.TOTL.ZS',   
    'SL.TLF.TOTL.IN',  

    'EN.ATM.CO2E.KT',  
    'EN.POP.DNST',      
    'EG.ELC.ACCS.ZS',   
    'IT.NET.USER.ZS',   
    'IC.LGR.DURS.US',   
    'IC.REG.COST.PC.ZS',

    'GC.DOD.TOTL.GD.ZS',
    'DT.ODA.ALLD.KD', 
    'SP.DYN.IMRT.IN',   
    'IQ.CPA.CONS.XQ',   
]
start_year = 1990
end_year = 2020
countries = 'all'
```

```python
try:
      df = wb.download(
          indicator=indicators,
          country=countries,
          start=start_year,
          end=end_year
      )
      df = df.reset_index()
      print("Data loaded successfully with 30+ features.")
except Exception as e:
      print(f"Error: {e}")
```

```python
column_mapping = {
    'NY.GDP.MKTP.CD': 'GDP_Total_USD',
    'NY.GDP.PCAP.CD': 'GDP_Per_Capita_USD',
    'TX.VAL.MRCH.CD.WT': 'Merchandise_Exports_USD',
    'NE.IMP.GNFS.ZS': 'Imports_Pct_GDP',
    'NV.IND.TOTL.ZS': 'Industry_Pct_GDP',
    'NV.AGR.TOTL.ZS': 'Agriculture_Pct_GDP',
    'NY.GNS.ICTR.ZS': 'Gross_Savings_Pct',
    'FP.CPI.TOTL.ZG': 'Inflation_Annual_Pct',
    'SP.POP.TOTL': 'Population_Total',
    'SP.URB.TOTL.IN.ZS': 'Urban_Population_Pct',
    'SM.POP.NETM': 'Net_Migration',
    'SH.DYN.MORT': 'Mortality_Under5_Rate',
    'SP.DYN.LE00.IN': 'Life_Expectancy_Years',
    'SH.STA.SUIC.P5': 'Suicide_Mortality_Rate',
    'SH.MLR.IMTR.ZS': 'Measles_Immunization_Pct',
    'SE.ADT.LITR.ZS': 'Literacy_Rate_Pct',
    'SE.PRM.ENRR': 'Primary_Enrollment_Pct',
    'SE.SEC.ENRR': 'Secondary_Enrollment_Pct',
    'SL.UEM.TOTL.ZS': 'Unemployment_Pct',
    'SL.TLF.TOTL.IN': 'Labor_Force_Total',
    'EN.ATM.CO2E.KT': 'CO2_Emissions_kT',
    'EN.POP.DNST': 'Population_Density',
    'EG.ELC.ACCS.ZS': 'Electricity_Access_Pct',
    'IT.NET.USER.ZS': 'Internet_Users_Pct',
    'IC.LGR.DURS.US': 'Time_to_Start_Business_Days',
    'IC.REG.COST.PC.ZS': 'Cost_to_Start_Business_Pct_GNI',
    'GC.DOD.TOTL.GD.ZS': 'Govt_Debt_Pct_GDP',
    'DT.ODA.ALLD.KD': 'Net_Aid_Received_USD',
    'SP.DYN.IMRT.IN': 'Infant_Mortality_Rate',
    'IQ.CPA.CONS.XQ': 'CPIA_Economic_Mgt_Rating',
}
df.rename(columns=column_mapping, inplace=True)
print(df.head())
```

```python
missing_pct = df.isnull().sum() / len(df) * 100
missing_df = pd.DataFrame({'Missing_Count': df.isnull().sum(), 'Missing_Percent': missing_pct})
high_missing_cols = missing_df[missing_df['Missing_Percent'] > 50].sort_values(by='Missing_Percent', ascending=False)
print(high_missing_cols)
```

```python
cols_to_drop = high_missing_cols[high_missing_cols['Missing_Percent'] > 70].index.tolist()
df.drop(columns=cols_to_drop, inplace=True)
print(f"\n--- {len(cols_to_drop)} Columns Dropped ---")
```

```python
print(df.columns)
```

```python
for col in df.select_dtypes(include=np.number).columns:
    if df[col].isnull().any():
        df[col] = df.groupby('country')[col].transform(lambda x: x.fillna(x.mean()))
df.fillna(df.median(numeric_only=True), inplace=True)
print(df.isnull().sum())
```

```python
df['year'] = df['year'].astype(int)
```

Global Trend (Average)

```python
global_trends = df.groupby('year')[['GDP_Total_USD', 'Internet_Users_Pct']].mean().reset_index()
print(global_trends.head())
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
sns.lineplot(x='year', y='GDP_Total_USD', data=global_trends, color='blue', ax=ax1, label='Average Global GDP (USD)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Average Global GDP (Trillions USD)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ax2 = ax1.twinx()
sns.lineplot(x='year', y='Internet_Users_Pct', data=global_trends, color='red', ax=ax2, label='Average Internet Users (%)')
ax2.set_ylabel('Average Internet Users (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
plt.title('Global Trend: GDP vs. Internet Access (1990-2020)')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
fig.tight_layout()
plt.show()
```

Correlation Matrix

```python
corr_cols = [
    'GDP_Per_Capita_USD',
    'Internet_Users_Pct',
    'Life_Expectancy_Years',
    'Electricity_Access_Pct',
    'Mortality_Under5_Rate'
]
correlation_matrix = df[corr_cols].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=.5,
    linecolor='black'
)
plt.title('Correlation Between Key Global Development Indicators (GDP, Health, Tech)')
plt.show()
```

Scatter Plots

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='Internet_Users_Pct',
    y='GDP_Per_Capita_USD',
    data=df,
    hue='year',
    palette='viridis',
    size='Population_Total',
    alpha=0.7
)
plt.title('Relationship: Internet Access vs. GDP Per Capita (1990-2020)')
plt.xlabel('Internet Users (% of Population)')
plt.ylabel('GDP Per Capita (USD)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Year')
plt.show()
```

Region Wise Analysis

```python
df_latest = df[df['year'] == 2020].copy()
bins = [0, 1000, 5000, 20000, df_latest['GDP_Per_Capita_USD'].max()]
labels = ['Low_Income', 'Lower_Middle_Income', 'Upper_Middle_Income', 'High_Income']
df_latest['Income_Group'] = pd.cut(df_latest['GDP_Per_Capita_USD'], bins=bins, labels=labels, right=False)
income_group_analysis = df_latest.groupby('Income_Group')['Life_Expectancy_Years'].mean().reset_index()
print(income_group_analysis)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Income_Group',
    y='Life_Expectancy_Years',
    data=income_group_analysis,
    palette='viridis'
)
plt.title('Life Expectancy (Years) by Income Group (2020 Data)')
plt.xlabel('Income Group (Based on GDP Per Capita)')
plt.ylabel('Average Life Expectancy (Years)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
